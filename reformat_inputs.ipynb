```python
xis = ['xi0', 'xi2', 'xi4']
dfs = []
for xi in xis:
    df = pd.read_csv(inpath + f'quadfits_LRG_FastPM_Y3_fnl_{xi}.csv', index_col=0)
    df = df.reset_index(names='s')
    # df['index'].rename = 's'
    df['term'] = len(df)*[xi]
    dfs.append(df)
df_out = pd.concat(dfs)
df_out = df_out.reset_index(drop=True)
df_out = df_out[['term', 's', 'c1', 'c2']]
df_out.to_csv('./inputs/quadfits_LRG_FastPM_Y3_fnl.csv',index=False)
```

```python
xis = ['xi0', 'xi2', 'xi4']
ksys_labels = ['SGC', 'DEC', 'MZLS']


for ksys in ksys_labels:  
    dfs = []
    for xi in xis:
        df = pd.read_csv(inpath + f'quadfits_LRG_abacus_Y1_Ksys_{ksys}_{xi}.csv', index_col=0)
        df = df.reset_index(names='s')
        # df['index'].rename = 's'
        df['term'] = len(df)*[xi]
        dfs.append(df)
    df_out = pd.concat(dfs)
    df_out = df_out.reset_index(drop=True)
    df_out = df_out[['term', 's', 'c1', 'c2']]
    df_out.to_csv(f'./inputs/quadfits_LRG_abacus_Y1_Ksys_{ksys}.csv',index=False)
```
